<h1> The tight-binding approximation </h1>

<p>
    The <a target="_tab" href=http://en.wikipedia.org/wiki/Tight_binding> tight binding</a> approximation is a method to calculate the electronic structure of condensed matter systems using <a target="_tab" href=https://en.wikipedia.org/wiki/Linear_combination_of_atomic_orbitals> linear combination of atomic orbitals</a> to represent the electronic wavefunctions. Most often, the elements of the tight binding hamiltonian are not computed directly but are rather fitted on data from other sources (experiments or more advanced calculations). In many situations, the resulting model provides qualitative results at a very low computational cost compared with <a target="_tab" href=https://en.wikipedia.org/wiki/Ab_initio_quantum_chemistry_methods>first-principles methods</a>.
</p>

<h3 style="color:#0B69A5;">
    Linear combination of atomic orbitals (LCAOs)
</h3>
<p>
    In order to derive the electronic spectrum of a quantum mechanical system, one has to solve the corresponding time-independent<a target="_tab" href=http://en.wikipedia.org/wiki/Schr%C3%B6dinger_equation> Schroedinger equation</a>,
</p>
<p class="math"> &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; $\hat{H}\, |\Psi \rangle= E \, |\Psi \rangle$, </p>
<p>
    where $\hat{H}$ is the <a target="_tab" href=http://en.wikipedia.org/wiki/Hamiltonian_%28quantum_mechanics%29> Hamiltonian </a> of the system. Within the tight binding approximation, the solutions to the time-independent single electron Schr√∂dinger equation are approximated as a linear combination of <a target="_tab" href=https://en.wikipedia.org/wiki/Atomic_orbital>atomic orbitals</a> $\{\phi(\vec{r})\}$, 
</p>
<p class="math">
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; $\Psi(\vec{r}) = \sum_{i,\alpha} \, c_{i}(\vec{R}_{\alpha}) \, \phi_{i}(\vec{r}-\vec{R}_{\alpha}) = \sum_{\mu} \, c_{\mu} \, \phi_{\mu}(\vec{r})$
</p>
<p>
    Here, $i$ refers to the i$^{th}$ atomic energy level, and $\vec{R}_{\alpha}$ locates the atom $\alpha$ in the crystal lattice. Without loosing in generality, the second equality introduces a combined indice $\mu$ that indexes the $(\alpha, i)$ pairs. 
</p>

<h3 style="color:#0B69A5;">
    The translational symmetry of the crystal
</h3>
<p>     
    In the case of electrons in a periodic potential, we know that the electronic eigenstates are <a target="_tab" href=http://en.wikipedia.org/wiki/Bloch_theorem> Bloch functions</a> :
</p>
<p class="math">
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;$\Psi(\vec{r}) = e^{i\vec{k}.\vec{r}}.u(\vec{r})$, 
</p>
<p>
    where $\vec{k}$ is the <a target="_tab" href=https://en.wikipedia.org/wiki/Wave_vector>wavevector</a> of the wavefunction and $u(\vec{r})$ has the periodicity of the potential. Hence, under translation over a lattice vector $\vec{R}_l$, the wave function only changes by a phase factor,
</p>
<p class="math">
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;$\Psi(\vec{r}+\vec{R}_l) = e^{i\vec{k}.\vec{R}_l}\Psi(\vec{r})$ .
</p>
<p>
    As a consequence, the coefficients $c_{i}(\vec{R}_{\alpha})$ satisfy the relation,
</p>
<p class="math">
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;$c_{i}(\vec{R}_{\alpha}+\vec{R}_{l}) = e^{i\vec{k}.\vec{R}_l} \, c_{i}(\vec{R}_{\alpha})$,
</p>
<p>
    and the electronic wavefunction can be conveniently rewritten as a finite sum over Bloch-like atomic orbitals,
</p>
<p class="math">
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;$\Psi(\vec{r}) = \sum_{\mu_0} \, c_{\mu_0}(\vec{k}) \, \Phi_{\mu_0}(\vec{r},\vec{k}) $, 
</p>
<p>
    where,
</p>
<p class="math">
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;$\Phi_{\mu_0}(\vec{r},\vec{k}) = \frac{1}{\sqrt{N}}\,\sum_{\mu' \equiv \mu_0}  e^{i\vec{k}.\vec{R}_{\mu'}} \phi(\vec{r}-\vec{R}_{\mu'}) $.
</p>
<p>
    Here, $\mu_0$ refers to atomic orbitals contained within the unit cell and $\sum_{\mu'\equiv \mu_0}$ denotes a sum running over all the atomic orbitals that are equivalent to $\phi(\vec{r}-\vec{R}_{\mu_0})$ by means of lattice vector translations. The prefactor $1/\sqrt{N}$ has been introduced so as to normalize the Bloch-like atomic orbitals. $N$ denotes the number of unit cells in the crystal. The approximated electronic wavefunctions are now expressed in a way that naturally accounts for the translational symmetry of the crystal. Each electronic wavefunction is characterized by a specific wavevector $\vec{k}$. In what follows, we will denote $E_j(\vec{k})$ and $\Psi_j(\vec{k})$, the j$^{th}$ electronic energy level at wavevector $\vec{k}$ and its associated electronic eigenstate.
</p>

<h3 style="color:#0B69A5;">
    The LCAOs eigenvalue equations
</h3>
<p>The stationary states of a quantum mechanical systems are the solutions of the time-independent<a target="_tab" href=http://en.wikipedia.org/wiki/Schr%C3%B6dinger_equation> Schroedinger equation</a>,</p>
<p class="math"> &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;$\hat{H}\, |\Psi_j(\vec{k}) \rangle = E_j(\vec{k})\, |\Psi_j(\vec{k})\rangle $ .</p>
<p>
    By substituting $\Psi_j(\vec{k})$ and projecting onto the set of Bloch-like atomic orbitals $\{ \Phi_{\mu_0}(\vec{k})\}$, one gets a <a target="_tab" href=http://en.wikipedia.org/wiki/Eigendecomposition_of_a_matrix#Generalized_eigenvalue_problem> generalized eigenvalue equation </a> that uniquely determines the expansion coefficients $\{ c_{j,\mu_0}(\vec{k}) \}$ of the electronic eigenstates characterized by the wavevector $\vec{k}$:
</p>
<p class="step">
    (a) Substitution :
</p>
<p class="math">
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;$ \frac{1}{\sqrt{N}} \,\sum_{\mu_0} \, c_{j,\mu_0}(\vec{k}) \, \hat{H}\,| \Phi_{\mu_0} \rangle = E_j(\vec{k})\, \frac{1}{\sqrt{N}} \,\sum_{\mu_0} \, c_{j,\mu_0}(\vec{k}) \, |\Phi_{\mu_0} \rangle$
</p>
<p class="math">
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;$ \frac{1}{\sqrt{N}} \,\sum_{\mu_0} \, c_{j,\mu_0}(\vec{k}) \, \sum_{\mu' \equiv \mu_0}  e^{i\vec{k}.\vec{R}_{\mu'}}\, \hat{H}\,| \phi_{\mu'} \rangle = E_j(\vec{k})\, \frac{1}{\sqrt{N}} \,\sum_{\mu_0} \, c_{j,\mu_0}(\vec{k}) \, \sum_{\mu' \equiv \mu_0}  e^{i\vec{k}.\vec{R}_{\mu'}}\, | \phi_{\mu'}\rangle$
</p>
<p class="step">
    (b) Projection :
</p>
<p class="math">
    $\frac{1}{N} \, \sum_{\mu_0} \, c_{j,\mu_0}(\vec{k}) \,  \sum_{(\nu' \equiv \nu_0, \mu' \equiv \mu_0)}\,  e^{i\vec{k}.(\vec{R}_{\mu'}-\vec{R}_{\nu'})}\, \langle \phi_{\nu'}\, | \hat{H}\,| \phi_{\mu'} \rangle = E_j(\vec{k})\, \frac{1}{N} \,\sum_{\mu_0} \, c_{j,\mu_0}(\vec{k}) \, \sum_{(\nu' \equiv \nu_0, \mu' \equiv \mu_0)}\, e^{i\vec{k}.(\vec{R}_{\mu'}-\vec{R}_{\nu'})}  \, \langle \phi_{\nu'}\, | \phi_{\mu'}\rangle$
</p>
<p class="step">
    (c) Generalized eigenvalue equation :
</p>
<p class="math">
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;$H_{\nu_0 \mu_0}(\vec{k}) \, c_{j,\mu_0}(\vec{k}) = E_j(\vec{k})\, S_{\nu_0 \mu_0}\, c_{j,\mu_0}(\vec{k})$
</p>
<p class="step">
    with
</p>
<p class="math">
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;$H_{\nu_0 \mu_0}(\vec{k}) = \langle \Phi_{\nu_0}(\vec{r},\vec{k}) |\hat{H} | \Phi_{\mu_0}(\vec{r},\vec{k}) \rangle = \frac{1}{N} \,
    \sum_{(\nu' \equiv \nu_0, \mu' \equiv \mu_0)}\,  e^{i\vec{k}.(\vec{R}_{\mu'}-\vec{R}_{\nu'})}\, \langle \phi_{\nu'}\, | \hat{H}\,| \phi_{\mu'} \rangle  = 
    \sum_{(\mu' \equiv \mu_0)}\,  e^{i\vec{k}.(\vec{R}_{\mu'}-\vec{R}_{\nu_0})}\, \langle \phi_{\nu_0}\, | \hat{H}\,| \phi_{\mu'}\rangle $
</p>
<p class="math">
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;$S_{\nu_0 \mu_0}(\vec{k}) = \langle \Phi_{\nu_0}(\vec{r},\vec{k}) | \Phi_{\mu_0}(\vec{r},\vec{k}) \rangle = \frac{1}{N} \,
    \sum_{(\nu' \equiv \nu_0, \mu' \equiv \mu_0)}\,  e^{i\vec{k}.(\vec{R}_{\mu'}-\vec{R}_{\nu'})}\, \langle \phi_{\nu'}\,| \phi_{\mu'} \rangle =
    \sum_{(\mu' \equiv \mu_0)}\,  e^{i\vec{k}.(\vec{R}_{\mu'}-\vec{R}_{\nu_0})}\, \langle \phi_{\nu_0}\,| \phi_{\mu'} \rangle
    $
</p>

<p>
   If we assume that the atomic orbitals $\{ \phi_{\mu}(\vec{r})\}$ are orthogonal, the equation further simplifies to the <a target="_tab" href=http://en.wikipedia.org/wiki/Eigenvalues_and_eigenvectors> eigenvalue problem </a>:
</p>
<p class="math">
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;$H_{\nu_0 \mu_0}(\vec{k}) \, c_{j,\mu_0}(\vec{k}) = E_j(\vec{k})\, c_{j,\mu_0}(\vec{k})$
</p>
<p>
    Solving this equation for a given wavevector $\vec{k}$ provides the energies and the expansion coefficients of the electronic wavefunctions $\Psi_j(\vec{k})$. Solving the equation for all the wavevectors within the first <a target="_tab" href=http://en.wikipedia.org/wiki/Brillouin_zone> Brillouin zone </a> fully determines the tight binding <a target="_tab" href=http://en.wikipedia.org/wiki/Electronic_band_structure> band structure </a> of the crystal.
</p>

<h3 style="color:#0B69A5;">
    Common tight-binding approximations
</h3>

<p>
    The elements of the LCAOs hamiltonian matrice have the general form :
</p>
<p class="math">
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;$H_{\nu_0 \mu_0}(\vec{k}) =  
    \sum_{(\mu' \equiv \mu_0)}\,  e^{i\vec{k}.(\vec{R}_{\mu'}-\vec{R}_{\nu_0})}\, \langle \phi_{\nu_0}\, | \hat{H}\,| \phi_{\mu'}\rangle =
    \sum_{(\mu' \equiv \mu_0)}\,  e^{i\vec{k}.(\vec{R}_{\mu'}-\vec{R}_{\nu_0})}\, \langle \phi_{\nu_0}\, | \sum_{\alpha} \left(\hat{H}^0_{\alpha} + \Delta\hat{U}_{\alpha} \right)\,| \phi_{\mu'}\rangle $
</p>
<p>
    Here, $\hat{H}^0_{\alpha}$ is the Hamiltonian of the isolated atom $\alpha$ and $\Delta\hat{U}_{\alpha}$ is the modification of the atomic Hamiltonian induced by the crystal. The diagonal elements of the LCAOs Hamiltonian are commonly named the "on-site energies" as they set the atomic energy levels at each sites. The out-of-diagonal elements are commonly called the "hopping integrals" as they are related to the probability of an electron to "jump" between sites. In practice, the matrice elements are not directly computed as we do not know a priori the operator $\Delta\hat{U}_{\alpha}$. A common strategy is to simplify the functional form of the Hamiltonian by keeping only the elements $\langle \phi_{\nu_0}\, | \hat{H}\,| \phi_{\mu'}\rangle$ that are expected to be most significant, and to treat these as parameters whose values are fitted on experimental data or on theoretical band structure calculations from other sources. 
</p>
<p>
    In practice, the most widely used approach to building tight-binding approximations consists in (a) selecting the smallest appropriate set of atomic orbitals to expand the wavefunctions of interest, and (b) considering only the interactions between neighbouring lattice sites. Along this road, many tight-binding calculations include only the interaction between the nearest-neighbours. If needed, the quality of the model can subsequently be improved by (a) adding atomic orbitals to the basis set used to expand the wavefunctions, and (b) increasing the number of neighbours entering the hamiltonian (second nearest-neighbours, third nearest-neighbours, and so on).
</p>

<h4 style="color:#0B69A5;">
    Note : Translational symmetry in reciprocal space
</h4>

    
<p>
    From the <a target="_tab" href=http://en.wikipedia.org/wiki/Bloch_wave> Bloch theorem </a>, we know that the eigenstates of an electron in a periodic potential can be written as Bloch waves: 
</p>
<p class="math">
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;$\Psi(\vec{r}) = e^{i\vec{k}.\vec{r}}.u(\vec{r})$. 
</p>

<div class="rblock">
    <img align='right' src='https://upload.wikimedia.org/wikipedia/commons/5/56/BlochWaves1D.svg' height="400" width="400"> 
</div>

<p>
    It is interesting to realize that if $\Psi(\vec{r})$ can be indexed by the wavevector $\vec{k}$, it can be equivalently indexed by the wavevector $\vec{k}+\vec{K}_l$, where $\vec{K}_l$ is any <a target="_tab" href=http://en.wikipedia.org/wiki/Reciprocal_lattice> reciprocal lattice vector </a>. This is illustrated in Fig. 1. A Bloch wave (bottom) can be broken up into the product of a periodic function (top) and a plane-wave (center). The left and right side represent the same Bloch wave broken up in two different ways, involving the wave vectors $\vec{k}_1$ (left) and $\vec{k}_2 +\vec{K}_l$ (right). Here blue and red corresponds to the real and imaginary part of the Bloch wave. 
</p>



<p>   
    Therefore, wavevectors whose difference corresponds to a reciprocal lattice vector are equivalent in the sense that they characterize the same subset of <a target="_tab" href=http://en.wikipedia.org/wiki/Bloch_wave> Bloch states </a>. To fully characterize the electronic band structure of a system, one can thus restrict the study to the wavevectors contained in the first <a target="_tab" href=http://en.wikipedia.org/wiki/Brillouin_zone> Brillouin zone</a>. By sake of convenience, the electronic band structure of crystals are often depicted along predefined paths within the first <a target="_tab" href=http://en.wikipedia.org/wiki/Brillouin_zone> Brillouin zone</a>. Commonly those paths are made of line segments connecting the point of high symmetry in reciprocal space. 
</p>

<h2> Exercise 1 : </br> The tight-binding model of graphene</h2>

<p>


<a target="_tab" href=http://en.wikipedia.org/wiki/Graphene>Graphene</a> is the two-dimensional allotropic form of carbon in which atoms are arranged in an hexagonal network. The  stability of the lattice is due to the strong  bond that forms in between the <a target="_tab" href=https://en.wikipedia.org/wiki/Orbital_hybridisation>$sp^2$ hybridized carbon atoms</a>. The $2s$, $2p_x$, $2p_y$ atomic orbitals of carbon take part in this hybridization process and are responsible for the $\sigma$-bonds. The eigenenergies of the associated bonding and anti-bonding bands lie far away from the <a target="_tab" href=http://en.wikipedia.org/wiki/Fermi_energy> Fermi level</a>, ($\delta E>3$eV). The low energy properties of graphene (e.g. its ultra-high electron mobility, its optical properties, etc...) are ruled by the remaining $2p_z$ atomic orbital of carbon that forms highly delocalized $\pi$-bonds. The separation of the energy range associated with the $\sigma$ and $\pi$ bands allows to describe the low energy physics of graphene with the $p_z$ atomic orbitals only. In what follows, we derive a tight-binding model of graphene within the so-called $\pi$-band approximation and we discuss its limitations. 
</p>

<h3>1.a The band structure of graphene with one orbital per site</h3>
<br>
In this exercise we use the tight-binding approximation to compute the band structure of graphene. The tight-binding model that we use here is very simple. We consider only one $p_z$ atomic orbital per carbon site and we restrict the interactions to the nearest-neighbours. We set the only parameter of the model, i.e. the hopping integral between neighbouring sites, to 2.6 eV. This value ensure a good agreement with experiments.
<br><br>
Questions :<br>
- What do you observe, is graphene metallic, semiconducting, or insulating? <br>
- What happens if you modify the onsite energy? Can you explain why?

```python
# Import the required python modules
import supertb as tb # use to carry out the tight-binding calculation
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize']=(8,6)

### Define the atoms
# create a species using tb.NumericalAtom(label, symbol, [orbitals])
carbon = tb.NumericalAtom('C','C',orbs=['pz'])
# create a table of elements using tb.AtomicTable(species1,species2,...)
atomictable = tb.AtomicTable(carbon)           

### Build the structure
# create a lattice using tb.Lattice.from_parameters(a,b,c,alpha,beta,gamma)
lattice = tb.Lattice.from_parameters(2.46,2.46,20.,90.,90.,60.)  
# create a list of fractional coordinates
pos = [[0.,0.,0.],[1./3,1./3,0.]]  
# build the structure using tb.Structure(lattice, list_of_species, list_of_coordinates)
struct = tb.Structure(lattice, ['C','C'], pos, coords_are_cartesian=False) 

### Define the interatomic interactions
# from the structure build a graph charaterized by a cutoff radius
graph = tb.StructureGraph.init_from_structure(struct, 1.45)    # 1.45 ist the c-c distance 
# define the kind of tight-binding model
sk = tb.SlaterKosterParams(atomictable, spinpol=False, spinorb=False)

### set the local tight-binding parameters
# set the Cp onsite energy
sk.set_local_integrals([('C', 0, 0, 0.)])
# set the Cp-Cp (pi-bond) hopping parameters
sk.set_non_local_integrals([('C','C', 0, 0, 3, 2.6)])  # 0 it the diagonal value and 2.6 is the energie value according to pi bonds

### Set up the tight binding model 
egraph = tb.ElectronicStructure(graph, atomictable, H=sk)
# and compute the hamiltonian
egraph.compute_integrals('H')

### Plot the band structure
# Define the high-symmetry path in the Brillouin zone
kbands = [[0.,0.,0.],[0.5,0.,0.],\
          [0.5,0.,0.],[2./3.,1./3.,0.],\     # Gamma, M , K, and K
          [2./3.,1./3.,0.],[0.,0.,0.]]
nkpts = [50,50,50]
# Diagonalize the hamiltonian at each k-point
bands = egraph.bands_structure(include=['H'], lines=kbands, nkpts=nkpts)
# Plot the bands structure
bands.plot_bands(xscale=True, klabels=['$\Gamma$','M','K','$\Gamma$'], \
                 vertices=[0,49,100,150], spincolors=['blue'],\
                 linestyle='-', emin=-8., emax=8.)

plt.show()


# The structure is a semimetal because of the slight overlap between the conduction and valence bands, semimetals have no band gap and 
#a negligible density of states at the Fermi level.

# to aswer to the second question, we use sk.set_local_integrals([('C', 0, 0, 2.)]). we'll see that, crossover point increases from 0.0 to 2.0 
```

<h3>1.b The band structure of graphene with four orbital per site</h3>
<br>
In this exercise we develop a more complete tight-binding model for graphene. This time we consider a set of four atomic orbitals per carbon site. We still restrict the interactions to the nearest-neighbours. Again, the values of the parameters are set to ensure a reasonable agreement with experiments.
<br><br>
Questions: <br>
- What do you observe? Can you locate the Dirac cones? There more bands, why?<br>
- Does this calculation validate the $\pi$-band approximation for the description of the low energy electronic properties of graphene?

```python
# Define the atoms
carbon = tb.NumericalAtom('C','C',orbs=['s','px','py','pz'])
atomictable = tb.AtomicTable(carbon)

# Build the structure
lattice = tb.Lattice.from_parameters(2.46,2.46,20.,90.,90.,60.)
pos = [[0.,0.,0.],[1./3,1./3,0.]]
struct = tb.Structure(lattice, ['C','C'], pos, coords_are_cartesian=False)

# Define the interatomic interactions
graph = tb.StructureGraph.init_from_structure(struct, 1.45)
sk = tb.SlaterKosterParams(atomictable, spinpol=False, spinorb=False)

# set the Cs onsite energy
sk.set_local_integrals([('C', 0, 0, -4.25956174899)])
# set the Cp onsite energy
sk.set_local_integrals([('C', 1, 1,  0.0476436867717)])
sk.set_local_integrals([('C', 2, 2,  0.0476436867717)])
sk.set_local_integrals([('C', 3, 3,  0.0476436867717)])
# set the Cs-Cs (sigma0-bond) hopping parameters
sk.set_non_local_integrals([('C', 'C', 0, 0, 0,  4.99016309759)])
# set the Cs-Cp (sigma1-bond) hopping parameters
sk.set_non_local_integrals([('C', 'C', 0, 0, 1,  4.93523388607)])
# set the Cp-Cp (sigma2-bond) hopping parameters
sk.set_non_local_integrals([('C', 'C', 0, 0, 2, -4.06687028938)])
# set the Cp-Cp (pi-bond) hopping parameters
sk.set_non_local_integrals([('C', 'C', 0, 0, 3,  2.61251108391)])


# Compute the interatomic integrals
egraph = tb.ElectronicStructure(graph, atomictable, H=sk)
egraph.compute_integrals('H')

# Define the high-symmetry path in the Brillouin zone
kbands = [[0.,0.,0.],[0.5,0.,0.],\
          [0.5,0.,0.],[2./3.,1./3.,0.],\
          [2./3.,1./3.,0.],[0.,0.,0.]]
nkpts = [50,50,50]
bands = egraph.bands_structure(include=['H'], lines=kbands, nkpts=nkpts)

# Plot the bands structure
bands.plot_bands(xscale=True, klabels=['$\Gamma$','M','K','$\Gamma$'], \
                 vertices=[0,49,100,150], spincolors=['blue'],linestyle='-')

plt.show()

# The dirac cones is at line 4, 
```

<h3>1.c The band structure of graphene including interactions up to the second-nearest neighbours</h3>
<br>
In this exercise we come back to our first model with one orbital per carbon site but we include interactions up to the second-nearest neighbours. The values of the parameters are set to ensure a reasonable agreement with experiments.
<br><br>
Question : <br>
- What is the main effect of the second nearest neighbours interaction?


```python
# Define the atoms
carbon = tb.NumericalAtom('C','C',orbs=['pz'])
atomictable = tb.AtomicTable(carbon)

# Build the structure
lattice = tb.Lattice.from_parameters(2.46,2.46,20.,90.,90.,60.)
pos = [[0.,0.,0.],[1./3,1./3,0.]]
struct = tb.Structure(lattice, ['C','C'], pos, coords_are_cartesian=False)

# Define the interatomic interactions
graph = tb.StructureGraph.init_from_structure(struct, 2.6)

def radial_func(hopping,length,hop1=0.,hop2=0.):
    if length<1.45:
        return hop1
    elif length<2.5:
        return hop2
    else:
        return 0.

sk = tb.SlaterKosterParams(atomictable, spinpol=False, spinorb=False, radialf=radial_func)
# set the Cp onsite energy
sk.set_local_integrals([('C', 0, 0, 0.)])
# set the Cp-Cp (pi-bond) hopping parameters for first and second nearest neighbors
sk.set_non_local_integrals([('C','C', 0, 0, 3, 0., {'hop2': 0.3, 'hop1': 2.6})])

# Compute the interatomic integrals
egraph = tb.ElectronicStructure(graph, atomictable, H=sk)
egraph.compute_integrals('H')

# Define the high-symmetry path in the Brillouin zone
kbands = [[0.,0.,0.],[0.5,0.,0.],\
          [0.5,0.,0.],[2./3.,1./3.,0.],\
          [2./3.,1./3.,0.],[0.,0.,0.]]
nkpts = [50,50,50]
bands = egraph.bands_structure(include=['H'], lines=kbands, nkpts=nkpts)

# Plot the bands structure
bands.plot_bands(xscale=True, klabels=['$\Gamma$','M','K','$\Gamma$'], \
                 vertices=[0,49,100,150], spincolors=['blue'],linestyle='-')

plt.show()
```

<h3>1.d The band structure of hexagonal Boron Nitride </h3>
<br>
In this exercise we adapt the model developped previously for graphene so as to get intuition about the band structure of hexagonal Boron Nitride (hBN). Hexagonal boron nitride has the same structure as graphene except that one of the atoms in the unit-cell is Boron and the other is Nitrogen. 
<br><br>

<h3 style="color:#0B69A5;"> Born-von Karman boundary condition</h3>
<p> 
    Neglecting the impact of the curvature, a carbon nanotube can be pictured as a planar portion of graphene with periodic boundary condition applied perpendicular to the axe of the tube. The consequence of the applied periodic boundary condition is that the component of the wavevector in that direction is quantized. Only given values of $\vec{k}_{perp}$ are compatible with the CNT structure. This is the well know <a target="_tab" href="http://en.wikipedia.org/wiki/Born%E2%80%93von_Karman_boundary_condition ">Born von Karman boundary condition</a>. The quantization of the allowed wavevectors of a nanotube strongly affects its electronic properties. One can show that a nanotube characterized by the (n,m) pair of indices, will be metallic if n = m, semiconducting with a very small gap if (n ‚àí m) is a multiple of 3, and semiconducting with a larger band gap, otherwise. This $(n-m=3j)$ rule is of geometrical nature and can be easily understood by means of the band folding approximation.  
</p>
